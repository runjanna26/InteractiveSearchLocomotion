```python
cpg = CPG_AFDC()
cpg_output_0 = []
cpg_output_1 = []
online_imitated_paths = {'TR0':[], 'CR0':[], 'FR0':[]}
for i in range(10000):
    

    if i < 1500:
        cpg.s = 1.0
        cpg.w00 = cpg.alpha * np.cos(0.05)
        cpg.w01 = cpg.alpha * np.sin(0.05)
        cpg.w10 = cpg.alpha * (-np.sin(0.05))
        cpg.w11 = cpg.alpha * np.cos(0.05)
        cpg.update_cpg_so2(0.0)      # phi would be equal when learning
    elif i > 1500:
        cpg.s = 1000.0
        cpg.w00 = 0.001
        cpg.w01 = 0 
        cpg.w10 = 0 
        cpg.w11 = 0.001 
        cpg.update_cpg_so2(0.0)


        
    cpg_output_0.append(cpg.out0_t)
    cpg_output_1.append(cpg.out1_t)
    
    
fig = go.Figure()
fig = make_subplots(rows=1, cols=1, shared_xaxes=False)
fig.update_layout(height=200, width=1100, title="CPG plot")
fig.update_layout(margin=dict(l=10, r=10, t=40, b=10))  # Reduce margins

fig.add_trace(go.Scatter( y=cpg_output_0, mode='lines', name=f'cpg_output_0'), row=1, col=1)
fig.add_trace(go.Scatter( y=cpg_output_1, mode='lines', name=f'cpg_output_1'), row=1, col=1)
```

```python
cpg = CPG_AFDC()
cpg_output_0 = []
cpg_output_1 = []
online_imitated_paths = {'TR0':[], 'CR0':[], 'FR0':[]}
random_number = np.random.randint(0, 501)
for i in range(500):
    if i < random_number:
        cpg.s = 1.0
        cpg.w00 = cpg.alpha * np.cos(0.05)
        cpg.w01 = cpg.alpha * np.sin(0.05)
        cpg.w10 = cpg.alpha * (-np.sin(0.05))
        cpg.w11 = cpg.alpha * np.cos(0.05)
        cpg.update_cpg_so2(0.0)      # phi would be equal when learning
        cpg_output_0.append(cpg.out0_t)
        cpg_output_1.append(cpg.out1_t)
    elif i >= random_number:
        cpg.s = 1.0
        cpg.w00 = cpg.alpha * np.cos(0.05)
        cpg.w01 = -cpg.alpha * np.sin(0.05)
        cpg.w10 = -cpg.alpha * (-np.sin(0.05))
        cpg.w11 = cpg.alpha * np.cos(0.05)
        cpg.update_cpg_so2(0.0)
        cpg_output_0.append(cpg.out0_t)
        cpg_output_1.append(cpg.out1_t)

        

    
    
fig = go.Figure()
fig = make_subplots(rows=1, cols=1, shared_xaxes=False)
fig.update_layout(height=200, width=1100, title="CPG plot")
fig.update_layout(margin=dict(l=10, r=10, t=40, b=10))  # Reduce margins

fig.add_trace(go.Scatter( y=cpg_output_0, mode='lines', name=f'cpg_output_0'), row=1, col=1)
fig.add_trace(go.Scatter( y=cpg_output_1, mode='lines', name=f'cpg_output_1'), row=1, col=1)
```
