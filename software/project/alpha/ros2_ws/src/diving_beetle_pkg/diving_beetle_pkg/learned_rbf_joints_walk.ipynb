```python
from scipy.interpolate import interp1d
# resample target
def resample_cycle(cycle, num_points):
    """Resamples a gait cycle to have a fixed number of points using interpolation."""
    original_indices = np.linspace(0, 1, len(cycle))  # original indices based on the length of the cycle
    new_indices = np.linspace(0, 1, num_points)  # new indices to match the desired number of points
    interpolation_function = interp1d(original_indices, cycle, kind='linear')  # interpolate the data
    return interpolation_function(new_indices)  # return the resampled cycle

def rotate_cyclic(df, column_name, range):
    # Move the last value to the first position using pd.concat()
    return pd.concat([df[column_name].iloc[-range:], df[column_name].iloc[:-range]]).reset_index(drop=True)
```

```python
support_offset = 0.4

stride_scale = 1.5
lift_scale = 1.5

height_offset_1_front = 0.3  # 0.8
height_offset_2_front = height_offset_1_front*1.4

height_offset_1_back = -0.6  # 0.8
height_offset_2_back = height_offset_1_back*1

leg_FR0_df = stride_scale * (stick_insect_walk['TR0']) + support_offset
leg_FR1_df = np.full(len(stick_insect_walk['TR0']), 0.0)  # -np.pi/9
leg_FR2_df = lift_scale * (stick_insect_walk['CR0']) - height_offset_1_front
leg_FR3_df = lift_scale * (stick_insect_walk['FR0']) + height_offset_2_front

leg_BR0_df = stride_scale * (stick_insect_walk['TR0']) - support_offset
leg_BR1_df = np.full(len(stick_insect_walk['TR0']), 0.0) # np.pi/9
leg_BR2_df = lift_scale * (stick_insect_walk['CR0']) - height_offset_1_back
leg_BR3_df = lift_scale * (stick_insect_walk['FR0']) + height_offset_2_back



leg_FL0_df = -leg_FR0_df.copy()
leg_FL1_df = np.full(len(stick_insect_walk['TR0']), 0.0) # np.pi/9
leg_FL2_df = leg_FR2_df.copy()
leg_FL3_df = leg_FR3_df.copy()

leg_BL0_df = -leg_BR0_df.copy()
leg_BL1_df = np.full(len(stick_insect_walk['TR0']), 0.0) # -np.pi/9
leg_BL2_df = leg_BR2_df.copy()
leg_BL3_df = leg_BR3_df.copy()


leg_BR = pd.DataFrame({'BR0': leg_BR0_df, 'BR1': leg_BR1_df, 'BR2': leg_BR2_df, 'BR3': leg_BR3_df})
leg_BL = pd.DataFrame({'BL0': leg_BL0_df, 'BL1': leg_BL1_df, 'BL2': leg_BL2_df, 'BL3': leg_BL3_df})
leg_FR = pd.DataFrame({'FR0': leg_FR0_df, 'FR1': leg_FR1_df, 'FR2': leg_FR2_df, 'FR3': leg_FR3_df})
leg_FL = pd.DataFrame({'FL0': leg_FL0_df, 'FL1': leg_FL1_df, 'FL2': leg_FL2_df, 'FL3': leg_FL3_df})
```

```python
from include.cpg_rbf.cpg_so2 import CPG_SO2
# from include.cpg_rbf.cpg_afdc import CPG_AFDC
from include.cpg_rbf.rbf import RBF

rbf = RBF()
cpg = CPG_SO2()
cpg_one_cycle = cpg.generate_cpg_one_cycle(0.05)    # lowest phi value
out0_cpg_one_cycle = cpg_one_cycle['out0_cpg_one_cycle'][:] # Use full cycle of CPG
out1_cpg_one_cycle = cpg_one_cycle['out1_cpg_one_cycle'][:] # Use full cycle of CPG
cpg_cycle_length = len(out0_cpg_one_cycle)
cpg_kernel = rbf.construct_kernels_with_cpg_one_cycle(out0_cpg_one_cycle, out1_cpg_one_cycle, cpg_cycle_length) # construct kernels with cpg one cycle
```

```python

# Define dataframes and corresponding column groups
dataframes = [leg_BR, leg_BL, leg_FR, leg_FL]
column_groups = [
    ['BR0', 'BR1', 'BR2', 'BR3'],
    ['BL0', 'BL1', 'BL2', 'BL3'],
    ['FR0', 'FR1', 'FR2', 'FR3'],
    ['FL0', 'FL1', 'FL2', 'FL3'],
]

imitated_paths = {}
imitated_weights = {}

# Loop through each dataframe and corresponding column group
for df, columns in zip(dataframes, column_groups):
    for col in columns:
        # Resample cycle for the current column
        target = resample_cycle(np.array(df[col]), cpg_cycle_length)
        
        # Imitate path and get weights using RBF
        imitated_paths[col] = rbf.imitate_path_by_learning(target)
        imitated_weights[col] = rbf.get_rbf_weight()
print(f"Saving weights for columns: {list(imitated_weights.keys())}")

np.savez('imitated_diving_beetle_walk_forward_weights.npz', **imitated_weights)
# np.savez('imitated_diving_beetle_swim_backward_weights.npz', **imitated_weights)
# np.savez('imitated_diving_beetle_swim_turning_weights.npz', **imitated_weights)
# np.savez('imitated_diving_beetle_swim_crossing_weights.npz', **imitated_weights)
```
